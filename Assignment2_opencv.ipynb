1. Perform bitwise XOR and not operation


```
import cv2
import numpy as np

rectangle=np.zeros((300,300),dtype="uint8")
cv2.rectangle(rectangle,(25,25),(275,275),(255,0,0),-1)
cv2_imshow(rectangle)

circle=np.zeros((300,300),dtype="uint8")
cv2.circle(circle,(150,150),150,(250,0,0),-1)
cv2_imshow(circle)

```

```
bitwisexor= cv2.bitwise_xor(rectangle, circle)
cv2_imshow(bitwisexor)
```

```
bitwiseNOT=cv2.bitwise_not(rectangle,circle)
cv2_imshow(bitwiseNOT)
```
